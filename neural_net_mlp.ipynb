```python
import math
import pandas as pd
import numpy as np
```

```python
test_df = pd.read_csv("WineQT_test.csv")
train_df = pd.read_csv("WineQT_train.csv")
```

```python
X_train = train_df.drop(['quality'], axis=1)
Y_train = pd.DataFrame(train_df['quality'])
X_test = test_df.drop(['quality'], axis=1)
Y_test = pd.DataFrame(test_df['quality'])
```

```python
X_train.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(X_train)
scaled_X_train = pd.DataFrame(scaler.transform(X_train))
scaler.fit(X_test)
scaled_X_test = pd.DataFrame(scaler.transform(X_test))
```

```python
scaled_X_train.head()
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
```

```python
for i in range(10):
  clf = MLPClassifier(hidden_layer_sizes=(i+1,),max_iter=1000).fit(scaled_X_train, Y_train.values.ravel());
  y_hat = clf.predict(scaled_X_test);
  cv_score = cross_val_score(clf, scaled_X_train, Y_train.values.ravel(), cv=10);
  accuracy = accuracy_score(Y_test, y_hat);
  print("Cross validation scores for ", i+1, " hidden layer(s): ", np.round(cv_score, 2))
  print("Mean cross validation scores for ", i+1, " hidden layer(s): ", np.round(np.mean(cv_score), 2))
  print("Accuracy score for ", i+1, " hidden layer(s): ", np.round(accuracy, 2))
```

As more nodes are used in the model, test scores seem to increase slightly for cross validation, and decrease when testing on the test data. This is an expected result as more model complexity improves model performance on training, but not testing data. As model complexity increases, the model goes from relatively high bias, low variance, to low bias, high variance.
